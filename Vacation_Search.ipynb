```python
# Import the dependencies.
import pandas as pd
import gmaps
import requests
# Import the API key.
from config import g_key
```

```python
# Store the CSV you saved created in part one into a DataFrame.
city_data_df = pd.read_csv("weather_data/WeatherPy_challenge.csv")
city_data_df.head(10)
```

```python
# Configure gmaps to use your Google API key.
gmaps.configure(api_key=g_key)
```

```python
# Ask the customer to add a minimum and maximum temperature value.
min_temp = float(input("What is the minimum temperature you would like for your trip? "))
max_temp = float(input("What is the maximum temperature you would like for your trip? "))
rain_details = input("Do you want it to be raining? (yes/no) ")
snow_details = input("Do you want it to be snowing? (yes/no) ")
```

```python
# Select the rain and snow value based on the user input
if (rain_details =="yes") and (snow_details == "yes"):
    rain_snow_user = (city_data_df["Rain Inches(last 3 hrs)"].gt(0.00)) | (city_data_df["Snow Inches(last 3 hrs)"].gt(0.00))
elif (rain_details =="yes") and (snow_details == "no"):
    rain_snow_user = (city_data_df["Rain Inches(last 3 hrs)"].gt(0.00)) & (city_data_df["Snow Inches(last 3 hrs)"].eq(0.00))
elif (rain_details =="no") and (snow_details == "yes"):
    rain_snow_user = (city_data_df["Rain Inches(last 3 hrs)"].eq(0.00)) & (city_data_df["Snow Inches(last 3 hrs)"].gt(0.00))
else:
    rain_snow_user = (city_data_df["Rain Inches(last 3 hrs)"].eq(0.00)) & (city_data_df["Snow Inches(last 3 hrs)"].eq(0.00))

# Create the temperature range based on the user input
max_temp_user = (city_data_df["Max Temp"] <= max_temp)
min_temp_user = (city_data_df["Max Temp"] >= min_temp)


# Select all the cities based on the user input.
preferred_cities_df = city_data_df[max_temp_user & min_temp_user & rain_snow_user]

preferred_cities_df.head(10)
```

```python
preferred_cities_df.count()
```

```python
# Create DataFrame called hotel_df to store hotel names along with city, country, max temp, and coordinates.
hotel_df = preferred_cities_df[["City", "Country", "Max Temp", "Lat", "Lng", "Current Description"]].copy()
hotel_df["Hotel Name"] = ""
hotel_df.head(10)
```

```python
# Set parameters to search for a hotel.
params = {
    "radius": 5000,
    "type": "lodging",
    "key": g_key
}
```

```python
# Iterate through the DataFrame.
for index, row in hotel_df.iterrows():
    # Get the latitude and longitude.
    lat = row["Lat"]
    lng = row["Lng"]

    # Add the latitude and longitude to location key for the params dictionary.
    params["location"] = f"{lat},{lng}"

    # Use the search term: "lodging" and our latitude and longitude.
    base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
    # Make request and get the JSON data from the search.
    hotels = requests.get(base_url, params=params).json()
    # Grab the first hotel from the results and store the name.
    try:
        hotel_df.loc[index, "Hotel Name"] = hotels["results"][0]["name"]
    except (IndexError):
        print("Hotel not found... skipping.")
```

```python
# Create the output file (CSV).
output_data_file = "weather_data/WeatherPy_vacation.csv"
# Export the hotel_Data into a CSV.
hotel_df.to_csv(output_data_file, index_label="City_ID")
```

```python
# Add a heatmap of temperature for the vacation spots.
locations = hotel_df[["Lat", "Lng"]]
max_temp = hotel_df["Max Temp"]
fig = gmaps.figure(center=(30.0, 31.0), zoom_level=1.5)
heat_layer = gmaps.heatmap_layer(locations, weights=max_temp, dissipating=False, max_intensity=300, point_radius=4)
marker_layer = gmaps.marker_layer(locations)
fig.add_layer(heat_layer)
fig.add_layer(marker_layer)
# Call the figure to plot the data.
fig
```

```python
info_box_template = """
<dl>
<dt>Hotel Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
<dt>Current Weather</dt><dd>{Current Description} and {Max Temp} °F </dd>
</dl>
"""
# Store the DataFrame Row.
hotel_info = [info_box_template.format(**row) for index, row in hotel_df.iterrows()]
```

```python
# Add a heatmap of temperature for the vacation spots and a pop-up marker for each city.
locations = hotel_df[["Lat", "Lng"]]
max_temp = hotel_df["Max Temp"]
fig = gmaps.figure(center=(30.0, 31.0), zoom_level=1.5)
heat_layer = gmaps.heatmap_layer(locations, weights=max_temp,dissipating=False,
             max_intensity=300, point_radius=4)
marker_layer = gmaps.marker_layer(locations, info_box_content=hotel_info)
fig.add_layer(heat_layer)
fig.add_layer(marker_layer)

# Call the figure to plot the data.
fig
```
